# outliers

as we have right skewed dataset , we will use IQR technique to handle outliers

```python
X.skew()
```

```python
# We have highly skewed dataset, Lets see boxplot
```

```python
for i in [x for x in X.columns if X[x].dtypes!='O']:
    sns.distplot(X[i])
    plt.show()
```

```python
for i in [x for x in X.columns if X[x].dtypes!='O']:
    sns.boxplot(X[i])
    plt.show()
```

```python
# some features are of categorical nature.
X.head()
```

```python
outlier=['aon',
 'daily_decr30',
 'daily_decr90',
 'rental30',
 'rental90',
 'last_rech_date_ma',
 'last_rech_date_da',
 'last_rech_amt_ma',
 'cnt_ma_rech30',
 'fr_ma_rech30',
 'sumamnt_ma_rech30',
 'medianamnt_ma_rech30',
 'medianmarechprebal30',
 'cnt_ma_rech90',
 'fr_ma_rech90',
 'sumamnt_ma_rech90',
 'medianamnt_ma_rech90',
 'medianmarechprebal90',
 'cnt_da_rech30',
 'fr_da_rech30',
 'cnt_da_rech90',
 'fr_da_rech90',
 'cnt_loans30',
 'amnt_loans30',
 'medianamnt_loans30',
 'cnt_loans90',
 'amnt_loans90',
 'maxamnt_loans90',
 'medianamnt_loans90',
 'payback30',
 'payback90',
  'day']
```

```python
for i in outlier:
    IQR= X[i].quantile(.75)-X[i].quantile(.25)
    lower=X[i].quantile(.25) - (1.5 * IQR)
    upper=X[i].quantile(.75) + (1.5 * IQR)
    X[i]=np.where(X[i]<lower,lower,X[i])
    X[i]=np.where(X[i]>upper,upper,X[i])
```

```python
X.skew()
```

```python
# Handling outliers has handled skewness in data
```

## Feature Selection through VIF

```python
# To find out VIF, we have to scale complete data first
cont=X.drop('month', axis=1)
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
scaler=sc.fit_transform(cont)

```

```python
scaler.shape
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
VIF=pd.DataFrame()
VIF['features']= cont.columns
VIF['VIF']=[variance_inflation_factor(scaler,i) for i in range(len(cont.columns))]
```
